```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrow

# Load the GeoJSON file for districts
geojson_path = './geodata/small_bangladesh_geojson_adm1_8_divisions_bibhags.json'
divisions = gpd.read_file(geojson_path)
```

```python
import pandas as pd
# load the data
data = pd.read_excel("./data/DENV_Regional_2019_2024.xlsx")

data = data.drop(columns=['Death'])

data
```

```python
df = data[0:8][["Division", "Infected"]]

division_data = df.groupby('Division').sum()
division_data
```

```python
df = data[0:8][["Division", "Infected"]]
division_data = df.groupby("Division").sum()
divisions["Infected"] = divisions["ADM1_EN"].map(division_data["Infected"]).fillna(0)

fig, ax = plt.subplots(figsize=(10, 10))
ax = divisions.plot(
    ax=ax,
    edgecolor="black",
    column="Infected",
    markersize=10,
    legend=True,
    cmap="tab20",
    legend_kwds={"label": "Infected Cases"},
)

divisions.apply(
    lambda x: ax.annotate(
        text=x.ADM1_EN,
        xy=x.geometry.centroid.coords[0],
        ha="center",
        fontweight="bold",
        color="black",
    ),
    axis=1,
)

ax.set_title("Dengue Infection Cases in Bangladesh 2024", fontsize=16)
```
